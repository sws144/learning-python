## Index and retrieving data

## Applying Function to Cells, Columns, Rows

````python
# apply function to each column (numpy's max)
df.apply(np.max)
````

````python
# can also apply to each row, using "lambda" to specify a new function and boolean index  (input:output)
df[df['State'].apply(lambda state: state[0] == 'W')].head()
````

````python
# map method is used to replace values in column by passing dictionary {old:new}
d = {'No' : False,'Yes' : True}
df['International plan'] = df['International plan'].map(d)
df.head()
````

````python
# same with replace method
df = df.replace({'Voice mail plan': d})
df.head()
````

## Grouping

In general, grouping works by: 
```
df.groupby(by=grouping_column)[columns_to_show].function()
```

````python
# groupby groups into rows
columns_to_show = ['Total day minutes', 'Total eve minutes', 'Total night minutes']

df.groupby(['Churn'])[columns_to_show].describe(percentiles=[])
````

````python
# same thing, but with agg() at the end for multiple single functions
df.groupby(['Churn'])[columns_to_show].agg([np.mean, np.std,np.min, np.max])
````

## Summary tables

````python
# build contingency table / crosstab
pd.crosstab(df['Churn'],df['International plan'])
````

````python
pd.crosstab(df['Churn'],df['Voice mail plan'], normalize=True)
````

````python
# use pivot_table
df.pivot_table(values=['Total day calls', 'Total eve calls', 'Total night calls'], index = ['Area code'], aggfunc = 'mean')
````

## DataFrame transformations

````python
# make new column for total calls for all users; this is a Series
total_calls =   df['Total day calls'] + df['Total eve calls'] + \
                df['Total night calls'] + df['Total intl calls']
df.insert(loc = len(df.columns), column = 'Total calls', value = total_calls)
# loc parameter is the number of columns after which to insert the Series object
# we set it to len(df.columns) to paste it at the very end of the dataframe
df.head()
````

````python
# add column directly
df['Total charge'] = df['Total day charge'] + df['Total eve charge'] + \
                     df['Total night charge'] + df['Total intl charge']
df.head()
````

````python
# delete columns using drop, axis = 1
df.drop(['Total charge','Total calls'], axis = 1, inplace =True)
df.drop([1,2]).head()
````

## First attempt at predicting telecom churn

````python
# use crosstab and seaborn
pd.crosstab(df['Churn'],df['International plan'], margins = True)
````

````python
# imports to set up plotting
import matplotlib.pyplot as plt
# pip install seaborn # if necessary
import seaborn as sns
# configure retina format for better graphics
%config InlineBackend.figure_format = 'retina'
````

````python
sns.countplot(x='International plan', hue = 'Churn', data=df)
````

Looks like international plans have higher churn (relative to group total)

````python
# customer service calls summary & picture
pd.crosstab(df['Churn'],df['Customer service calls'], margins =True)
````

````python
sns.countplot(x='Customer service calls', hue = 'Churn', data = df)
````

churn rates seems to increase as a % of those number of calls as # of calls increases

````python
# test if many service calls (>3) leads to more churn
df['Many_service_calls'] = (df['Customer service calls'] > 3).astype('int')

pd.crosstab(df['Many_service_calls'], df['Churn'], margins = True)
````

````python
sns.countplot(x='Many_service_calls',hue='Churn',data=df)
````

````python
# combine both many service & international plan metrics
TwoFeatureTable = pd.crosstab(df['Many_service_calls'] & df['International plan'] ,  df['Churn'])
print(TwoFeatureTable)
type(TwoFeatureTable)
````

````python
# baseline amount 
ChurnYes = df['Churn'].value_counts()[1]
TotalNumbers = df['Churn'].value_counts().sum()

#NumberCorrect from many service & intl plan
NumberCorrect =np.diag(TwoFeatureTable).sum()

print("Churnbaseline % = ChurnNo / TotalNumbers =  {0: 8.2f}%".format((1-ChurnYes/TotalNumbers)*100))
print("Accuracy % = NumberCorrect / TotalNumbers  = {0: 8.2f}%".format(NumberCorrect/TotalNumbers*100))
````

Need to beat at least this baseline to count churn
